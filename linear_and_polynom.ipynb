```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
df_str=pd.read_csv("sea-surface.csv")
```

```python
x= df_str["Year"]
y = df_str["Annual anomaly"]
```

```python
x=x.values.reshape(141,1)
y=y.values.reshape(141,1)
```

```python
plt.scatter(x,y)
```

```python
lin=LinearRegression()
lin.fit(x,y)
```

```python
lin.predict(x)
plt.scatter(x,y)
plt.plot(x, lin.predict(x), color="red")
```

```python
pol=PolynomialFeatures(degree=3)
newx=pol.fit_transform(x)
polmod=LinearRegression()
polmod.fit(newx,y)
polmod.predict(newx)
plt.scatter(x,y)
plt.plot(x,polmod.predict(newx),color="red")
```

```python
err=0
err2=0
for i in range(len(newx)):
    err2=err2+(float(y [i])-float(polmod.predict(newx)[i]))**2
for i in range (len(y)):
    err=err+(float(y [i])-float(lin.predict(x)[i]))**2
err2
```

```python
for a in range(90):
    pol=PolynomialFeatures(degree=a)
    newx=pol.fit_transform(x)
    polmod=LinearRegression()
    polmod.fit(newx,y)
    polmod.predict(newx)
    for i in range(len(newx)):
        err2=0
        err2=err2+(float(y [i])-float(polmod.predict(newx)[i]))**2
    print(a+1," value error",err2)
```

```python
pol=PolynomialFeatures(degree=85)
newx=pol.fit_transform(x)
polmod=LinearRegression()
polmod.fit(newx,y)
polmod.predict(newx)
plt.scatter(x,y)
plt.plot(x, lin.predict(x), color="red")
plt.plot(x,polmod.predict(newx),color="red")
```
